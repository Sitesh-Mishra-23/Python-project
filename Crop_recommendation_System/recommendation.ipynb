```python
import numpy as np
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


df = pd.read_csv("Crop_recommendation.csv")

X = df.drop("label", axis=1)
y = df["label"]


le = LabelEncoder()
y_encoded = le.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"Train: {X_train.shape[0]} samples  |  Test: {X_test.shape[0]} samples\n")


rf_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model",  RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1))
])

rf_pipeline.fit(X_train, y_train)
rf_acc = accuracy_score(y_test, rf_pipeline.predict(X_test))
print(f"Random Forest  →  Test Accuracy: {rf_acc:.4f}")


lr_pipeline = Pipeline([
    ("scaler", StandardScaler()),          
    ("model",  LogisticRegression(max_iter=1000, random_state=42))
])

lr_pipeline.fit(X_train, y_train)
lr_acc = accuracy_score(y_test, lr_pipeline.predict(X_test))
print(f"Logistic Reg   →  Test Accuracy: {lr_acc:.4f}\n")


for name, pipe in [("Random Forest", rf_pipeline), ("Logistic Regression", lr_pipeline)]:
    print(f"── {name} ──")
    print(classification_report(y_test, pipe.predict(X_test), target_names=le.classes_))



joblib.dump({"model": rf_pipeline, "label_encoder": le,
             "features": X.columns.tolist()}, "rf_crop_model.pkl")

joblib.dump({"model": lr_pipeline, "label_encoder": le,
             "features": X.columns.tolist()}, "lr_crop_model.pkl")

print("Saved:  rf_crop_model.pkl")
print("Saved:  lr_crop_model.pkl")

```
